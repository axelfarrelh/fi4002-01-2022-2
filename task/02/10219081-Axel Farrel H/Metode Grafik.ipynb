```python
#102190881-Axel Farrel Hutagalung
#Metode Grafik.

#Import Library
import matplotlib.pyplot as plt
import math

#definisikan fungsi f(x)
def f(x):
    y = math.pow(x, 2) - (10*x) + 5
    return y

x = [i for i in range(-5,5)]
y = [f(i) for i in x]


xroot = [1]
yroot = [0]

plt.grid()
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```

```python
xroot = [0.6]
yroot = [0]

plt.grid()
plt.xlim(-1, 1.5)
plt.ylim(-20, 20)
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```
