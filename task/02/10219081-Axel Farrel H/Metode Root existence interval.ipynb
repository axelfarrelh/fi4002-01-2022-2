```python
#10219081-Axel Farrel Hutagalung

#Import Library
import math
import matplotlib.pyplot as plt

# Definisikan Fungsi
def f(x):
    y = math.pow(x, 2) - (10*x) + 5
    return y

x = [i for i in range(-2,3)]
y = [f(i) for i in x]

xroot = [0, 1]
yroot = [f(i) for i in xroot]

plt.grid()
plt.plot(x, y, 'b-', xroot, yroot, 'ro', markerfacecolor='none', ms=10, lw=2)
plt.show()

print("xroot =", xroot)
print("yroot =", yroot)
print("f(x1) f(x2) =", yroot[0] * yroot[1], "< -1")
print("Root existence:", True)
```
